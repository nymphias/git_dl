```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
import mp1
%matplotlib auto
```

```python
from keras.models import Sequential
from keras.optimizers import SGD, Adam
from keras.utils import np_utils
from keras.layers.convolutional import ZeroPadding2D
from keras.layers import Dropout, Flatten, BatchNormalization, UpSampling2D, Conv2D, MaxPooling2D, Dense, Activation
```

```python
IMAGE_SIZE = 72
```

```python
def print_results(model, title, X_test = None, Y_test = None):
    """Plots accuracy and loss of the model during training and their value on test set

    Args:
        model: Net model
        title1: title to first graph (loss)
        title2: title to second graph (accuracy)
        (X_test, Y_test): test set
        
    Returns:
        Void

    """
    keys = model.history.history.keys()
    metrics = [key.replace('val_', '') for key in keys if key.startswith('val_')]
    f, ax = plt.subplots(len(metrics), 1, figsize = (15,12))
    
    for i, metric in enumerate(metrics):
        data_plot = [model.history.history[metric], model.history.history['val_'+metric]]
        mp1.plot_standard(data_plot, ['train', 'test'], None, title[i], metric, 'epochs', ax[i])
        
    if X_test is not None and Y_test is not None:
        values = model.evaluate(X_test, Y_test)
        print(tabulate(np.stack([metrics, values], axis = 1), headers=['metrics', 'values'], tablefmt='orgtbl'))
```

## 3- Simple Classification

__Question__: build and train a linear classifier to classify a image into three possible categories (rectangle, disk, triangle) where each shape center is fixed to the center of the image.

#### Generate train and test set where shapes are centered 

```python
nb_train = 300
[X_train, Y_train] = mp1.generate_dataset_classification(nb_train, 20, False)
# Dummy variable for categorical output
Y_train = np_utils.to_categorical(Y_train, 3)
```

```python
[X_test, Y_test] = mp1.generate_dataset_classification(nb_train, 20, False)
# Dummy variable for categorical output
Y_test = np_utils.to_categorical(Y_test, 3)
```

```python
# Setting number of epochs
epochs = 150
```

#### Linear classifier with softmax layer - SGD optimization (with momentum)


```python
model = Sequential()
model.add(Dense(3, input_shape = (IMAGE_SIZE**2,),activation = 'softmax'))

sgd = SGD(lr = 0.001, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer = sgd, metrics = ['accuracy'])
model.fit(X_train, Y_train, epochs = epochs, batch_size = 32, verbose = 0, validation_data = (X_test, Y_test))

title1 = 'Cross-entropy Loss - SGD optimization'
title2 = 'Model accuracy - SGD optimization'
print_results(model, [title1, title2], X_test, Y_test)
```

#### Linear classifier with softmax layer - ADAM optimization

```python
model = Sequential()
model.add(Dense(3, input_shape = (IMAGE_SIZE**2,),activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train, Y_train, batch_size = 32, epochs = epochs, verbose = 0, validation_data =  (X_test, Y_test))

title1 = 'Cross-entropy Loss - ADAM optimization'
title2 = 'Model accuracy - ADAM optimization'
print_results(model, [title1, title2], X_test, Y_test)
```

__Comments__:
For classification problem, we use a softmax layer to compute the conditional probabilities of each category. The loss function is the cross entropy function. We use two differents optimizers:
- SGD with momentum. Tuning of learning rate is important, it is stuck to a local minimum for lr = 0.01. For 0.001, the model accuracy is 1 on both test and train sets after about 40 epochs, with only nb_train=300.
- ADAM: It requires less tuning, and handles the strategy itself in updating direction and step size using first two moments. It converges in less epochs compared to SGD.

For both algorithms, the accuracy stabilizes first and the loss continues to decrease. In fact, accuracy does not depend on how high the softmax output is, just craes if the highest softmax output is the correct category one.

## 5- A More Difficult Classification Problem

#### Generate train and test set where shapes are no longer centered 

```python
nb_train = 1000
[X_train, Y_train] = mp1.generate_dataset_classification(nb_train, 20, True)

Y_train = np_utils.to_categorical(Y_train, 3)
```

```python
[X_test, Y_test] = mp1.generate_test_set_classification()
```

```python
epochs = 150
```

#### Linear classifier with softmax layer - ADAM optimization

```python
model = Sequential()
model.add(Dense(3, input_shape = (IMAGE_SIZE**2,),activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train, Y_train, batch_size = 32, epochs = epochs, verbose = 0, validation_data = (X_test,Y_test))

title1 = 'Cross-entropy Loss - ADAM optimization'
title2 = 'Model accuracy - ADAM optimization'
print_results(model, [title1, title2], X_test, Y_test)
```

#### ConvNet - ADAM optimization

```python
# Reshape images into matrices instead of 1d-vector
X_train = X_train.reshape(nb_train, IMAGE_SIZE, IMAGE_SIZE, 1)
X_test = X_test.reshape(len(X_test), IMAGE_SIZE, IMAGE_SIZE, 1)

model = Sequential()
# The kernel size defines the granularity of the local properties we want to extract
model.add(Conv2D(16, (5,5), input_shape = (IMAGE_SIZE, IMAGE_SIZE, 1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(3, activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train, Y_train, batch_size = 32, epochs = epochs, verbose = 0, validation_data = (X_test,Y_test))

title1 = 'Cross-entropy Loss - ADAM optimization'
title2 = 'Model accuracy - ADAM optimization'
print_results(model, [title1, title2], X_test, Y_test)
```

__Comments__:

This time, the shapes are no longer centered. It basically add 2 degrees of freedom: the center location and the rotation. 

- The simple classifier model doesn't perform as well as before with a test accuracy of 0.6 for nb_train = 1000 (it is still better than random guess where we would score a third). It also appears that we need more data as the profiles of accuracy and loss curves during training are quite noisy.
- We suggest a more complex model using ConvNet with 1 conv layer with a 5x5 kernel, 1 max-pooling layer and 1 dense layer. This model brings more stable results with a __test accuracy of 0.85__ and __loss function of 0.6 for nb_train = 1000__. The metrics are better for higher nb_train.

## 6- A Regression Problem

#### Generate train and test set for regression problem


```python
nb_train = 1000
```

```python
[X_train, Y_train] = mp1.generate_dataset_regression(nb_train, 20)
X_train = X_train.reshape(nb_train, IMAGE_SIZE, IMAGE_SIZE, 1)
# We sort the vertices along x-axis
Y_train = (mp1.sort_Y)(Y_train)
```

```python
[X_test, Y_test] = mp1.generate_test_set_regression()
X_test = X_test.reshape(len(X_test), IMAGE_SIZE, IMAGE_SIZE, 1)
# We sort the vertices along x-axis
Y_test = (mp1.sort_Y)(Y_test)
```

```python
epochs = 40
```

#### ConvNet with dropout - ADAM optimization

```python
model = Sequential()

model.add(Conv2D(64, (3,3), input_shape = (IMAGE_SIZE, IMAGE_SIZE, 1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(32, (2,2), input_shape = (IMAGE_SIZE, IMAGE_SIZE, 1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(6))

model.compile(loss='mean_squared_error', optimizer='adam', metrics = [mp1.r_square])
model.fit(X_train,Y_train,epochs=epochs,batch_size=32,validation_data=(X_test,Y_test), verbose = False)

title1 = 'Mean-Squared Loss - ADAM optimization'
title2 = 'Mean-Squared Loss - ADAM optimization'
print_results(model, [title1, title2], X_test, Y_test)
```

```python
idx = np.random.randint(len(X_test))
mp1.visualize_prediction(X_test[idx], model.predict(X_test)[idx])
```

__Comments__:

We consider now a regression problem where we want to find the vertices of a given triangle. As generate_a_triangle outputs the vertices in a arbitrary order, it is not reasonable to ask our model to guess that order. Hence, we need to sort the vertices along an axis: we choose to sort along x-axis.

Our model use a two Conv-layers with Pooling each time and dropout, with a final dense layer.
As it is a regression problem, we follow r2 metric and MSE loss. We obtain a __r2=0.7 and MSE=0.026 for nb_train=1000__. 

## 7- Image denoising

#### Generate train and test set for denoising problem


```python
nb_train = 300

[X_train, Y_train] = mp1.generate_dataset_denoise(nb_train, noise_minmax = [10,40])
X_train = X_train.reshape(nb_train, IMAGE_SIZE, IMAGE_SIZE, 1)
Y_train = Y_train.reshape(nb_train, IMAGE_SIZE, IMAGE_SIZE, 1)


[X_test, Y_test] = mp1.generate_test_set_denoise()
X_test = X_test.reshape(len(X_test), IMAGE_SIZE, IMAGE_SIZE, 1)
Y_test = Y_test.reshape(len(Y_test), IMAGE_SIZE, IMAGE_SIZE, 1)

```

#### Hourglass autoencoder model

```python
model =Sequential()

#Encoders
model.add(Conv2D(64,kernel_size=(3,3),input_shape=(IMAGE_SIZE,IMAGE_SIZE,1,),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(64,kernel_size=(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
print(model.output_shape)

model.add(Conv2D(128,kernel_size=(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(128,kernel_size=(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
print(model.output_shape)

#Decoders
model.add(UpSampling2D(size=(2,2)))
model.add(Conv2D(128,kernel_size=(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(128,kernel_size=(3,3),padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
print(model.output_shape)


model.add(ZeroPadding2D(padding=(1, 1)))
model.add(Conv2D(1,kernel_size=(3,3)))
model.add(Dense(1,activation='sigmoid'))
print(model.output_shape)

```

```python
epochs = 5
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, validation_data = [X_test, Y_test], epochs = epochs, batch_size = 100, verbose = 0)
title1 = 'Binary cross-entropy Loss - ADAM optimization'
title2 = 'Model accuracy - ADAM optimization'
print_results(model, [title1, title2], X_test, Y_test)
```

__Comments__:

We consider now a the problem of denoising which goal is to reconstruct corrupted data. We can allow more hidden features than input features without the risk to learn the idendity function as we input on purpose noisy data and when we compute the loss function, we compare the output values with the unnoisy input.

The suggested hourglass is a sequence of encoders (Pooling) and decoders (UpSampler) - we assure that input and onput dimensions are the same.

We see the problem as a binary classification one. In fact, the wanted output is a denoised image where we want that pixel with information = 1 and pixel without information = 0. Hence our denoised inputs follow this format. It is more straightforward to see this problem as a binary classification one than a regression.

The results are quite satisfying, we obtain __a loss of 0.0257416 and accuracy of 0.999115.__
